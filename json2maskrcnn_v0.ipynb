```python
# JSON 파일 불러와서 변수에 저장하기
import json 

JsonDataSets =['one_01', 'one_02']

for (i, dt) in enumerate(JsonDataSets):
    file = open(f"{dt}.json","r",encoding='utf-8')
    globals()['json_'+ str(i)] = json.load(file)
```

두개의 json 합치기

```python
# 전체 JSON 하나로 합치기
json_list = []
for (i, dt) in enumerate(JsonDataSets):
    json_list.append(globals()['json_'+ str(i)])
```
